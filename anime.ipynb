```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel("Anime.xlsx")
```

```python
df.dropna(how="all")
```

```python
name=df["Name"].head(10)
rank=df["Rank"].head(10)
```

```python
fig = plt.figure(figsize =(10, 7))
plt.bar(name[0:10], rank[0:10])
plt.title("Top 10 Anime")
plt.xlabel("Anime names")
plt.ylabel("Ratings")
plt.xticks(name,rotation="vertical",size=8)
plt.show()
```
